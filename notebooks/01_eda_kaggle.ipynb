# Exploratory Data Analysis (EDA)
## NASA Near-Earth Objects (Kaggle Dataset)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/neo_kaggle.csv")
df.head()
,
```

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns={
    'Is Potentially Hazardous': 'hazardous',
    'Absolute Magnitude (H)': 'absolute_magnitude_h',
    'Min Diameter (m)': 'estimated_diameter_min',
    'Max Diameter (m)': 'estimated_diameter_max',
    'Relative Velocity (km/s)': 'relative_velocity',
    'Miss Distance (astronomical)': 'miss_distance'
}, inplace=True)



```

```python
df['hazardous'].value_counts()
```

```python
features = [
    'absolute_magnitude_h',
    'estimated_diameter_min',
    'estimated_diameter_max',
    'relative_velocity',
    'miss_distance',
    'hazardous'
]

df = df[features]
df.head()
```

```python
df.isnull().sum()

```

```python
df = df.dropna()
```

```python
sns.countplot(x='hazardous', data=df)
plt.title("Hazardous vs Non-Hazardous Asteroids")
plt.show()
```

```python
df.to_csv("../data/processed/neo_clean.csv", index=False)
```
